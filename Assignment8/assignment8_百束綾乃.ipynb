## 課題2

以下の図を再現してください

- 男性乗客と女性乗客の年齢の分布をそれぞれ描画する
- 全乗客の年齢分布を示す灰色ヒストグラムも描画する
- 図のタイトル、x軸ラベル、y軸ラベル、凡例を設定する

![](./Figure2.png)

```python
age = df.age
df2 = df[df["sex"] == "male"]
df3 = df[df["sex"] == "female"]
age2 = df2.age
age3 = df3.age

fig,axs = plt.subplots(figsize=(10,8))
axs.hist(age, bins=50, color='gray')
axs.hist(age2, bins=50, color='blue')
axs.legend(["すべての乗客","男性乗客"])
axs.set_title("男性乗客")
axs.set_xlabel('年齢(歳)')
axs.set_ylabel('人数')
axs.set_yticks([0,10,20,30,40,50,60,70,80])
axs.grid(axis="y")
```
